# WEBSCRAPING CON CODIGO DE PYTHON

Usaremos las paginas web de las librerias online:<br>
-sbs libreria internacional (para Peru): https://www.sbs.com.pe<br>
-libreria Crisol: https://www.crisol.com.pe

## Proceso

Usaremos una serie de codigos en python que nos permitira busqueda del libro de nuestro interes, identificar todos los resultados 
y extraer los datos importantes de cada resultado.<br>
Luego haremos una comparacion de precios y disponibilidad entre ambas paginas web, esto con el fin de encontrar el mejor precio, el resultado lo guardaremos en un archivo csv.<br>
A continuacion explicaremos brevemente la funcion principal de cada trozo de codigo:<br>

### Funcion para extrer los datos

1. **def extraer_datos(driver, wait, busqueda, url_base, tienda_nombre)**:<br>
Esta función encapsula el motor de extracción de datos, operando en dos fases críticas para transformar una búsqueda web en información estructurada. Inicialmente, utiliza Selenium para automatizar la interacción con el buscador, enviando la consulta y ejecutando un comando de JavaScript para realizar un desplazamiento de página (scroll) que garantiza la carga de todos los elementos dinámicos. Tras estabilizar el DOM, captura los enlaces únicos de cada libro para evitar errores de referencias obsoletas (stale elements) y comienza un bucle iterativo de navegación profunda. Dentro de cada producto, BeautifulSoup analiza el código fuente para extraer el título, el precio y la ficha técnica desde tablas específicas, aplicando una limpieza de caracteres en el campo ISBN para normalizarlo. La disponibilidad se valida en tiempo real mediante la detección del botón de compra, y los precios se convierten a formato decimal para permitir comparaciones numéricas posteriores. El uso de bloques try-except asegura que el programa sea resiliente, permitiendo saltar errores individuales sin interrumpir el flujo general del raspado. Finalmente, todos los atributos recolectados se consolidan en una lista de diccionarios que servirá como base para los reportes y el análisis comparativo final.

2. **Segunda parte del codigo (flujo principal):**<br>
El flujo principal del script orquesta la lógica de negocio mediante la ejecución secuencial de tareas de extracción, procesamiento y 
comparación de datos bibliográficos. Primero, solicita al usuario el nombre del libro e inicializa el navegador configurado junto con 
un sistema de espera inteligente para gestionar la latencia de la red. Seguidamente, invoca la función de extracción para SBS y Crisol 
de forma independiente, transformando las listas de diccionarios obtenidas en objetos DataFrame de Pandas para facilitar su manipulación.
El código valida que existan resultados antes de imprimir tablas formateadas en consola y exportar los datos crudos a archivos CSV con 
codificación compatible para Excel. Para la fase analítica, realiza una limpieza de registros y ejecuta un cruce de datos
(merge) para encontrar coincidencias exactas de la misma edición en ambas tiendas. Dentro de un bucle iterativo, evalúa los precios 
numéricos para determinar algorítmicamente qué librería ofrece la mejor oferta económica o si existe un empate. Finalmente, construye 
un reporte consolidado con las etiquetas de "Mejor Opción", lo visualiza en una tabla estética de alta legibilidad y cierra de forma 
segura la instancia del navegador en el bloque finally para liberar recursos del sistema.



## Explicacion del resultado y conclusion

En el proceso de scrapear estas dos paginas web se uso como ejemplo el libro **Pinocho**, el cual se introdujo como palabra de busqueda 
en el codigo, el cual dio como resultado una tabla con los libros encontrados para cada libreria, estas tablas tenian detalles como el 
autor, la disponibilidad(en stock o agotado), el precio y el año de publicacion, luego se guardaron en formato csv y se hizo un cuadro
comparativo entre estas dos opciones para analizar que libro esta a un mejor precio.<br>
Para lograr que los libros comparados sean los mismos en ambas paginas se uso un codigo de indetificacion unico de cada libro 
llamado **ISBN** el cual reconocia aquellos libros que era iguales en las dos paginas, lo cual ayudaba en el momento de seleccion para su posterio comparacion.<br>
En conclusion se logro con exito el webscraping de estas paginas usando **BeautifulSoup** y **selenium**, ademas de usar la conversion 
de html a json para su correcta ejecusion, se logro mostrar y guardar las tablas de libros encontrados junto a datos relevantes de 
estos.<br>
El usuario puede ingresar el libro de su preferencia y usar nuestro algoritmo para una mejor toma de decisiones.
